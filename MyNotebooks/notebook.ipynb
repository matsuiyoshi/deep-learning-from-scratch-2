# 準備

`!python` と `%run` の違いメモ

- `!python script.py` : **別プロセスとして実行**される。ノートブックとは独立して動くため、`script.py` 内で作られた変数や関数はノートブック側には残らない。
- `%run script.py` : **同じカーネル内で実行**される。`script.py` 内で定義された変数や関数・クラスは、実行後にノートブックの後続セルからそのまま利用できる。

各コードセルは **1ファイル＝1セル** の `%run` です。

- パスは `MyNotebooks` をカレントにしたときの `../chNN/ファイル名.py` です。
- プロジェクト直下でカーネルを動かしている場合は `../` を外し、`chNN/...` にしてください。
